```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
sns.set()
```

Use the given iris.csv dataset to answer below questions
1. Matplotlib

```python
iris_data = pd.read_csv('Iris.csv')
iris_data.head()
```

1.1 
Plot the sepal length of the Iris dataset against the corresponding index values

```python
plt.figure()
plt.plot(iris_data['Sepal_Length'])
plt.xlabel('index')
plt.ylabel('sepal length')
plt.title('Sepal Length')
```

1.2 Create a scatter plot to plot the sepal length and width of the Iris dataset against each
other

```python
plt.figure()
plt.scatter(iris_data['Sepal_Length'],iris_data['Sepal_Width'])
plt.xlabel('Sepal_Length')
plt.ylabel('Sepal_Width')
plt.title('Sepal_Width VS Sepal_Length')
```

1.3 Using a bar graph plot the number of flowers of each species in the Iris dataset

```python
plt.figure()
plt.bar(iris_data['Species'].value_counts().index, iris_data['Species'].value_counts())
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('number of flowers of each species')
```

1.4 Create two plots side by side, one for the sepal length and width, and the other for the
petal length and width

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1).scatter(iris_data['Sepal_Length'],iris_data['Sepal_Width'])
plt.xlabel('Sepal_Length')
plt.ylabel('Sepal_Width')
plt.title('Sepal_Width VS Sepal_Length')

plt.subplot(1,2,2).scatter(iris_data['Petal_Length'],iris_data['Petal_Width'])
plt.xlabel('Petal_Length')
plt.ylabel('Petal_Width')
plt.title('Petal_Width VS Petal_Length')
```

1.5 How to add text labels to a (or to the above scatter plot) plot?
1.6 How to add annotations to the above plot?
1.7 How to adjust the font size, color, and style?
1.8 How to customize the default style and appearance of Matplotlib plots?
1.9 How to change the font, colors, and markers of a plot?
1.10 How to save a plot to a file in different formats, such as PNG, PDF, and SVG

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1).scatter(iris_data['Sepal_Length'],iris_data['Sepal_Width'],marker='o',color='purple')
plt.xlabel('Sepal_Length', color ='green')
plt.ylabel('Sepal_Width', color ='green')
plt.title('Sepal_Width VS Sepal_Length', size = 20,color='red')

plt.subplot(1,2,2).scatter(iris_data['Petal_Length'],iris_data['Petal_Width'],marker='x')
plt.xlabel('Petal_Length', color ='green')
plt.ylabel('Petal_Width', color ='green')
plt.title('Petal_Width VS Petal_Length', size = 20, color = 'red')

plt.suptitle("Scatter plot of Iris data", size = 25, style = 'italic', fontfamily='Times New Roman')

plt.savefig("activity3Q1.png")
plt.show()
```

Use the world happiness.csv dataset to answer below questions\
2. Pandas

```python
happiness_data = pd.read_csv('world-happiness.csv')
happiness_data.head()
```

2.1 Use the world happiness.csv dataset to answer below questions

```python
happiness_data = happiness_data.drop_duplicates('Country')
happiness_data.describe()
```

2.2 Fill in missing values using the mean value of the column

```python
happiness_data.notnull().sum()
```

```python
happiness_data['Generosity'].fillna(happiness_data['Generosity'].mean(),inplace=True)
happiness_data.notnull().sum()
```

2.3 How can you create a line plot of the Happiness Score column in the World Happiness
dataset using Pandas?

```python
plt.figure()
happiness_data['Happiness Score'].plot(kind='line')
plt.xlabel('Index')
plt.ylabel('Happiness Score')
plt.title('Happiness Score')
plt.show()
```

2.4 How can you create a bar plot of the average Happiness Score by Region in the World
Happiness dataset using Pandas

```python
plt.figure()
happiness_data.groupby('Region')['Happiness Score'].mean().plot(kind='bar')
plt.ylabel('Happiness Score Average')
plt.title('Happiness Score by Region')
plt.show()
```

2.5 How can you create a scatter plot of the Happiness Score column against the Economy
(GDP per Capita) column in the World Happiness dataset using Pandas?

```python
plt.figure()
happiness_data.plot(kind='scatter',x='Economy (GDP per Capita)',y='Happiness Score')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.title('Happiness Score column against the Economy(GDP per Capita) column')
plt.show()
```

2.6 Box plot and a Violin plot of happiness score by region

```python

plt.figure(figsize=(8, 6))  # Create a single figure with a specified size

# Boxplot

happiness_data.boxplot(column='Happiness Score', by='Region')
plt.xticks(rotation=45, ha='right')
plt.title('Happiness Score by Region (Boxplot)')  # Add title to the subplot

# Violin plot

grouped_data = happiness_data.groupby('Region')['Happiness Score'].apply(list)
plt.violinplot(grouped_data, showmeans=False, showextrema=False)
plt.xticks(range(1, len(grouped_data) + 1), grouped_data.index, rotation=45, ha='right')
plt.title('Happiness Score by Region (Violin Plot)')  # Add title to the subplot

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

2.7 Heatmap of correlations between variables

```python
happiness_data_corr = happiness_data.select_dtypes(include=['int','float'])
plt.imshow(happiness_data_corr.corr(),cmap='autumn',aspect='auto')
plt.colorbar(format ='%.2f')
plt.xticks(range(len(happiness_data_corr.columns)),happiness_data_corr.columns, rotation='vertical')
plt.yticks(range(len(happiness_data_corr.columns)),happiness_data_corr.columns)
plt.show()
```
