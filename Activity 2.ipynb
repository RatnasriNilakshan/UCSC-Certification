## Question 1

```python
#Q1 :Use Pandas to read the "countries of the world.csv" file and display the first five rows of the DataFrame.

import pandas as pd
df = pd.read_csv('countries of the world.csv')
df.head()
```

```python
#Q2 :Filter the rows of the DataFrame to show only the countries with a population greater than 100 million

df_filtered = df[df['Population'] > 100000000]
df_filtered
```

```python
# Q3 :filter the DataFrame to show only the name, region, and population columns

df_filtered = df[['Country', 'Region', 'Population']]
df_filtered.head()
#df_filtered
```

```python
# Q4 :sort the country details DataFrame in descending order based on population

df_sorted = df.sort_values(by=['Population'], ascending=False)
df_sorted.head()
```

```python
#Q5 :Rename the "Pop. Density (per sq. mi.)" column to "Population Density" and the "GDP ($ per capita)" column to "GDP" in the DataFrame

df_renamed = df.rename(columns={'Pop. Density (per sq. mi.)': 'Population Density', 'GDP ($ per capita)': 'GDP'})
df_renamed.head()
```

```python
# Q6 :group the DataFrame by the "Region" column and calculate the maximum population of each region

df_grouped = df.groupby(['Region'])['Population'].max()
df_grouped
```

```python
# Q7 :read the "Countries-Latitude & Longitude" file and load it into a DataFrame

df2 = pd.read_csv('Geo coordinates.csv')

# Remove whitespaces from the 'Country' column
df['Country'] = df['Country'].str.strip()

# merge the "countries of the world.csv" file with the "Geo coordinates.csv" file on the "Country" column
df_merged = pd.merge(df, df2, how='outer', on="Country")
df_merged.head()

# Merge Options:
# 'inner': Includes data where countries are present in both datasets.
# 'outer': Combines all country data, merging if available, leaving blanks where no match exists.
# 'left': Includes all country data from the "countries of the world.csv" dataset, merging if available.
# 'right': Includes all country data from the "Geo coordinates.csv" dataset, merging if available.
```

```python
#Q1 :check for missing values in the DataFrame
df.isnull().sum()

#df.isnull()
```

```python
#Q2 : calculate the percentage of missing values in each column of the DataFrame

percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing
```

```python
#Q3: drop the rows from the DataFrame that have missing values in the "Agriculture" column

df_dropped = df.dropna(subset=['Agriculture'])
df_dropped.head()
```

```python
#Q4 :fill the missing values in the "Climate" column with the mode value

df_filled = df.copy()
df_filled['Climate'].fillna(df_filled['Climate'].mode()[0], inplace=True)
df_filled.head(10)
```

```python
#Q6 : drop the duplicate rows from the DataFrame
df_dropped = df.drop_duplicates()
df_dropped
```

```python
#Q7 :Replace the value "EASTERN EUROPE" in the "Region" column with "EUROPE (EASTERN)"?

df_replaced = df.copy()
df_replaced['Region'] = df_replaced['Region'].replace('EASTERN EUROPE', 'EUROPE (EASTERN)')
df_replaced
```

```python
#Q8 Use a lambda function to create a new column in the DataFrame called Population Category  that categorizes countries based on the â€˜populationâ€™ as below.
#High: population greater than 100 million
#Medium: population between 10 million and 100 million
#Low: population less than 10 million

df_category = df.copy()
df_category['Population Category'] = df['Population'].apply(lambda x: 'High' if x > 100000000 else 'Medium' if x > 10000000 else 'Low')

df_category
```
