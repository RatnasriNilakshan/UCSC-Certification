Q1. Answer the following questions using the “countries of the world.csv” and “geo
coordinates.csv” files

1. Use Pandas to read the "countries of the world.csv" file and display the first five
rows of the DataFrame.

```python
import pandas as pd
```

```python
country_data = pd.read_csv("countries of the world.csv")
country_data.head()
```

2. Filter the rows of the DataFrame to show only the countries with a population
greater than 100 million

```python
highPopulation_countries = country_data[country_data['Population']>100000000]
highPopulation_countries
```

3. Filter the columns of the DataFrame to show only the name, region, and
population columns

```python
keyCountry_data = country_data[['Country','Region','Population']]
keyCountry_data
```

4. Sort the country details DataFrame in descending order based on Population.

```python
population_country_data = country_data.sort_values(by='Population', ascending=False)
population_country_data
```

5. Rename the "Pop. Density (per sq. mi.)" column to "Population Density" and the
"GDP ($ per capita)" column to "GDP" in the DataFrame

```python
renamed_data = country_data.rename(columns={"Pop. Density (per sq. mi.)" : "Population Density", "GDP ($ per capita)":"GDP"})
renamed_data.head()
```

6. Group the DataFrame by the "Region" column and get the maximum population
values of each region

```python
grouped_country_data = renamed_data.groupby("Region")["Population"].max()
grouped_country_data
```

7. Merge the "countries of the world.csv" file with the "geo coordinates.csv" file
and load it into a DataFrame

```python
renamed_data["Country"] = renamed_data["Country"].str.strip().astype(str)
geo_coordinates_data = pd.read_csv("geo coordinates.csv")
geo_coordinates_data_drop_na = geo_coordinates_data.drop_duplicates()
geo_coordinates_data_drop_na.loc[:, "Country"] = geo_coordinates_data_drop_na["Country"].str.strip().astype(str)
merged_data = pd.merge(renamed_data, geo_coordinates_data_drop_na, on="Country", how="left")
merged_data
```

Q2. Use the “countries of the world.csv” dataset to answer the following questions

1. How can we check if there are any missing values in the DataFrame?

```python
merged_data.isnull().sum()
```

2. Identify the percentages of missing values in each column of the DataFrame

```python
print("Missing values percentage according to the each coloumn \n")
for column_name in merged_data.columns:
    print(column_name,":",round(merged_data[column_name].isnull().sum()/len(merged_data["Country"])*100,2),"%")
```

3. Drop the rows from the DataFrame that have missing values in the "Agriculture"
column

```python
dp_agri_merged_data = merged_data.dropna(subset=["Agriculture"])
dp_agri_merged_data
```

4. Fill the missing values in the "Climate" column with the mode value

```python
mode_climate = dp_agri_merged_data["Climate"].mode()[0]
dp_agri_merged_data.loc[:, "Climate"] = dp_agri_merged_data["Climate"].fillna(mode_climate)
dp_agri_merged_data
```

5. Identify the number of duplicate rows in the DataFrame

```python
dp_agri_merged_data.duplicated().sum()
```

6. Drop the above identified duplicate rows from the DataFrame

```python
dp_agri_dp_merged_data = dp_agri_merged_data.drop_duplicates()
dp_agri_dp_merged_data
```

7. Replace the value "EASTERN EUROPE" in the "Region" column with "EUROPE
(EASTERN)"

```python
dp_agri_dp_merged_data.loc[:,"Region"] = dp_agri_dp_merged_data["Region"].replace("EASTERN EUROPE", "EUROPE (EASTERN)")
dp_agri_dp_merged_data.head()
```

8. Use a lambda function to create a new column in the DataFrame called
‘Population Category’, that categorizes countries based on the ’population’ as
below. \
●High: population greater than 100 million \
●Medium: population between 10 million and 100 million \
●Low: population less than 10 million
1


```python
categorize_population = lambda x: 'High' if x > 100000000 else ('Medium' if 10000000 <= x <= 100000000 else 'Low')
dp_agri_dp_merged_data.loc[:,'Population Category'] = dp_agri_dp_merged_data.loc[:,'Population'].apply(categorize_population)
dp_agri_dp_merged_data.head()
```
