```python
import pandas as pd
import numpy as np
from sklearn import svm , datasets
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
dataset = datasets.load_iris()
```

```python
X = dataset.data[:,:2]
y = dataset.target

```

```python
models = (svm.SVC(kernel='linear'),svm.LinearSVC(max_iter=10000),svm.SVC(kernel='rbf', gamma=0.7),svm.SVC(kernel='poly',degree = 3, gamma='auto'))
models = (clf.fit(X,y) for clf in models)
```

```python
titles = ('SVC with liner kernal', 'LinearSVC (liner model)', 'SVC with RBF kernal', 'SVC with polynomial (degree 3) kernal ')
```

```python
fig, sub = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:,1]

for clf, title, ax in zip (models, titles, sub.flatten()):
    disp = DecisionBoundaryDisplay.from_estimator(clf, X,response_method='predict',ax=ax, 
                                                  cmap = plt.cm.coolwarm, xlabel=dataset.feature_names[0], ylabel=dataset.feature_names[1])
    ax.scatter(X0,X1, c=y, cmap=plt.cm.coolwarm, s =20, edgecolors='k')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)
plt.show()
```
